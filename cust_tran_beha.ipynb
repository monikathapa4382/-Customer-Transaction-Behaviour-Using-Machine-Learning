```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('Santander_Customer_Transaction_Data.csv')
```

```python
data.sample(10)
```

```python
data.drop("ID_code", axis = 1, inplace = True)
```

```python
data.sample(10)
```

```python
print('There are', data.shape[0] ,' rows and', data.shape[1],' columns')
```

```python
any_missing = False
for column in data.columns:
    missing_values = data[column].isnull().sum()
    if missing_values > 0:
        print(f"Missing values in column '{column}': {missing_values}")
        any_missing = True

if not any_missing:
    print("No missing values in any column")
```

```python
print("Total number of duplicate rows in the dataset:", data.duplicated().sum())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
class_counts = data['target'].value_counts()
axes[0].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
axes[0].set_title("Percentage of Classes")
sns.countplot(data=data, x='target', ax=axes[1], palette="pastel")
axes[1].set_title("Count of Classes")
axes[1].set_xlabel("Target Class")
axes[1].set_ylabel("Count")
plt.tight_layout()
plt.show()
```

The objective of this challenge is to predict which customers are likely to complete a specific transaction in the future, regardless of the transaction amount. This involves building a model that can effectively distinguish between customers who will make the transaction (class 1) and those who will not (class 0). However, the dataset is highly imbalanced, with approximately 90% of the instances labeled as class 0 and only 10% as class 1. Such an imbalance can make it challenging for standard machine learning models to accurately identify the minority class, as they tend to favor the majority class in imbalanced datasets. Addressing this imbalance is essential to improve model performance and ensure that it can reliably predict the minority class, which is the group of customers likely to complete the transaction.

#### Distribution of Random Variables:

##### var_0 -

```python
plt.figure(figsize = (10,8))
var_0_0 = data[data['target'] == 0][data.columns[1]]
var_0_1 = data[data['target'] == 1][data.columns[1]]
plt.hist(var_0_0, bins=30, alpha=0.5, color='blue', label='Class 0', density=True)
plt.hist(var_0_1, bins=30, alpha=0.5, color='orange', label='Class 1', density=True)
sns.kdeplot(var_0_0, color='blue', label='Class 0', lw=2, linestyle='--')
sns.kdeplot(var_0_1, color='orange', label='Class 1', lw=2, linestyle='--')
plt.legend(title="Target Class")
plt.xlabel(data.columns[1])
plt.ylabel("Density")
plt.title("Distribution of Variable 0 by Target Class")
plt.show()
```
