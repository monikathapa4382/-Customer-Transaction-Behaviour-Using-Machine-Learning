The objective of this challenge is to predict which customers are likely to complete a specific transaction in the future, regardless of the transaction amount. This involves building a model that can effectively distinguish between customers who will make the transaction (class 1) and those who will not (class 0). However, the dataset is highly imbalanced, with approximately 90% of the instances labeled as class 0 and only 10% as class 1. Such an imbalance can make it challenging for standard machine learning models to accurately identify the minority class, as they tend to favor the majority class in imbalanced datasets. Addressing this imbalance is essential to improve model performance and ensure that it can reliably predict the minority class, which is the group of customers likely to complete the transaction.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
features = data.drop(columns='target')
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
pca = PCA()
pca.fit(features_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), cumulative_explained_variance, marker='o', color='blue', linestyle='--')
plt.title('Scree Plot (Cumulative Explained Variance)', fontsize=14)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=175)
pca_features = pca.fit_transform(features_scaled)
pca_df = pd.DataFrame(pca_features, columns=[f'PC{i+1}' for i in range(175)])
pca_df['target'] = data['target'].reset_index(drop=True)
pca_df.head()
```

```python
from sklearn.model_selection import train_test_split

f = pca_df.drop(columns = 'target', axis = 1)
t = pca_df['target']

x_train, x_test, y_train, y_test = train_test_split(f, t, test_size=0.2)
```

## Machine Learning

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
index = 0

n_estimators = [5, 10, 20, 25, 30, 50]

while index < len(n_estimators):
    i = n_estimators[index]
    rf = RandomForestClassifier(n_estimators=i)
    rf.fit(x_train, y_train)
    pred = rf.predict(x_train)
    acc = accuracy_score(y_train, pred)
    print(f'\nTrain accuracy for n_estimators = {i} is {acc:.2f}%')
    index += 1
```

```python
rf = RandomForestClassifier(n_estimators=5)
rf.fit(x_train, y_train)
pred = rf.predict(x_test)
acc = accuracy_score(y_test, pred)
print('Test accuracy for n_estimators = 5 is %f%%' % (acc))
```

```python
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 1', 'Class 2'], 
            yticklabels=['Class 1', 'Class 2'], cbar=False)
plt.title('Confusion Matrix for Random Forest Classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

### Decision Tree

```python
index = 0

max_depths = [5, 10, 15, 20, 25, 30, 50]

while index < len(max_depths):
    depth = max_depths[index]
    dt = DecisionTreeClassifier(max_depth=depth)
    dt.fit(x_train, y_train)
    pred = dt.predict(x_train)
    acc = accuracy_score(y_train, pred)
    print(f'\nTrain accuracy for max_depth = {depth} is {acc:.2f}%')
    index += 1
```

```python
dt = DecisionTreeClassifier(max_depth=50)
dt.fit(x_train, y_train)
pred = dt.predict(x_test)
acc = accuracy_score(y_test, pred)
print('Test accuracy for max_depth = 50 is %f%%' % (acc))
```

```python
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 1', 'Class 2'], 
            yticklabels=['Class 1', 'Class 2'], cbar=False)
plt.title('Confusion Matrix for Decision Tree Classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

### Logistic Regression

```python
index = 0

C_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

while index < len(C_values):
    C = C_values[index]
    log_reg = LogisticRegression(C=C, max_iter=1000)
    log_reg.fit(x_train, y_train)
    pred = log_reg.predict(x_train)
    acc = accuracy_score(y_train, pred)
    print(f'\nTrain accuracy for C = {C} is {acc:.2f}%')
    index += 1
```

```python
log_reg = LogisticRegression(C=1, max_iter=1000)
log_reg.fit(x_train, y_train)
pred = log_reg.predict(x_test)
acc = accuracy_score(y_test, pred)
print('Test accuracy for C = 1 is %f%%' % (acc))
```

```python
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 1', 'Class 2'], 
            yticklabels=['Class 1', 'Class 2'], cbar=False)
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

### XGBoost

```python
n_estimators_values = [10, 20, 50, 70, 100]

index = 0

while index < len(n_estimators_values):
    n_estimators = n_estimators_values[index]
    xgb_model = xgb.XGBClassifier(n_estimators=n_estimators, use_label_encoder=False, eval_metric='logloss')
    xgb_model.fit(x_train, y_train)
    pred = xgb_model.predict(x_train)
    acc = accuracy_score(y_train, pred)
    print(f'\nTrain accuracy for n_estimators = {n_estimators} is {acc:.2f}%')
    index += 1
```

```python
xgb_model = xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(x_train, y_train)
pred = xgb_model.predict(x_test)
acc = accuracy_score(y_test, pred)
print('Test accuracy for n_estimators = 100 is %f%%' % (acc))
```

```python
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 1', 'Class 2'], 
            yticklabels=['Class 1', 'Class 2'], cbar=False)
plt.title('Confusion Matrix for XGBoost Classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

## Deep Learning

### Multi-Layered Perceptron - HOLLOW

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
model = Sequential()

model.add(Dense(128, input_dim=f.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',  loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))
```

```python
y_pred_prob = model.predict(x_test)
y_pred = (y_pred_prob > 0.5).astype(int)
cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.show()
```

```python
history_dict = history.history
loss = history_dict['loss']
val_loss = history_dict['val_loss']
accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

epochs = range(1, len(loss) + 1)

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, label='Training Loss', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', marker='x')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, accuracy, label='Training Accuracy', marker='o')
plt.plot(epochs, val_accuracy, label='Validation Accuracy', marker='x')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

### Multi-Layered Perceptron - Dense

```python
model = Sequential()

model.add(Dense(128, input_dim=f.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',  loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))
```

```python
y_pred_prob = model.predict(x_test)
y_pred = (y_pred_prob > 0.5).astype(int)
cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.show()
```
